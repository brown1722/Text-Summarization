Test đọc 1 file

```python
import os

def txt_file_url(path_doc):
    txt = []
    for name_file in os.listdir(path_doc):
        path_txt = os.path.join(path_doc, name_file)
        if os.path.isfile(path_txt) and name_file.endswith(".txt"):
            txt.append(path_txt)
    return txt

def read_content(txt_file):
    with open(txt_file, 'r', encoding='utf-8') as f:
        content = f.read()
    return content

path_doc = 'D:/Data/NLP/ViMs/ViMs/original/Cluster_001/original/'
list_txt = txt_file_url(path_doc)

# Đọc từng file và in ra nội dung
for txt_file in list_txt:
    print(f"Nội dung của file {txt_file}:")
    noi_dung = read_content(txt_file)
    print(noi_dung)
    print("=" * 30)
```

Đọc dữ liệu, lưu vào csv

```python

def read_data(file_path):
    #input: a text file
    #output: title, Summary, Content
    with open(file_path,encoding='utf-8-sig') as f:
        title=f.readline()[7:]
        rows=f.readlines()
        summary=''
        content=''
        start_copy_summary=False
        start_copy_content=False

        for row in rows:
            if row[:7] == 'Content':
                start_copy_summary = False
                start_copy_content = True
            elif start_copy_content:
                content += row + ' '
            elif start_copy_summary:
                summary += row + ' '
            elif row[:7] == 'Summary': 
                start_copy_summary = True
                summary += row[9:] +' ' 
    return title,summary,content
```

```python
#Get_data
df=[]

for cluster_path in cluster_paths:
    list_txt = txt_file_url(cluster_path)
    for txt in list_txt:
        title, summary, content= read_data(txt)
        df.append({'Title' : title,'Summary':summary, 'Content':content})
        data=pd.DataFrame(df)

data.head(10)
```

```python
data.to_csv('data.csv', encoding='utf-8-sig', index = False)
```
